## Experiment 4 now i want thing to be optional and value to be more annotated

```python
from typing import List,Dict,Optional,Annotated
from pydantic import BaseModel,Field,EmailStr,AnyUrl
#Pydantic provides another data type also for validation , including email(it should have @ sign etc ), url(ending wiwth proper .com/in/eu etc) etc 
# field function will allow you to put the constraint, metadata and default value
#Annotated it helps to defines , first parameter should be data type, then field value
class NewIntro(BaseModel):
    name:Annotated[str,Field(max_length=50,title='Name of the person')] # Title will be visible in APIs
    age:int=Field(ge=0,Lt=120) #Using field we can specify other constraint rather just data type
    hobby:Annotated[Optional[List[str]],Field(default=None,max_length=5)] # it says if you provide int(or anything other than string) in the list , it will throw error
    contact_details:Optional[Dict[str,str]]=None # Optional 
    email:EmailStr


```

```python
def newintro(intro=NewIntro):
    print(intro.name)
    print(intro.age)
    print(intro.hobby)
    print(intro.contact_details)
    print(intro.email)

para={'name':'Deepak','age':30,'hobby':['cricket','badminton','chess'],'contact_details':{'mob':'1234','phn':'5678'}
      ,'email':'dwiv2011@gmail.com'}

newintro(NewIntro(**para))
```
